```python
%matplotlib inline
import matplotlib
from matplotlib import pyplot as plt
matplotlib.rcParams.update( {'font.size': 24})
```

```python
import numpy as np
```

```python
figs = dict()
```

```python
def one_more_today_than_tomorrow():
    plt.xkcd()
    x = np.array([ 1, 2 ])
    w = 0.8
    days = 5
    errorbar = 2
    fig, ax = plt.subplots()
    
    ax.bar( x, [ days, days - 1], w, yerr=[ errorbar, errorbar ])
    l = x[1] - x[0] - w
    m = np.mean(x)
    ax.plot( [ m - 0.5*l, m + 0.5*l], [days, days], 'k--')
    ax.plot( [ m - 0.5*l, m + 0.5*l], [days - 1, days - 1], 'k--')
    ax.text( m, days - 0.5, '1 day', ha='left', va='center', fontsize='large')
    ax.set_xticks( x)
    ax.set_xticklabels( ['Today', 'Tomorrow'], rotation=30, fontsize='large')
    ax.set_yticklabels( [])
    ax.set_ylabel( 'Number of days lived', fontsize='large')
    return fig

figs['one more today'] = one_more_today_than_tomorrow()
figs['one more today'].savefig( 'one-more-today-than-tomorrow.png', bbox_inches='tight', dpi=300)
```
